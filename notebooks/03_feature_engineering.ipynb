```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/bengaluru_house_prices.csv")
```

```python
df.head()
df.shape
df.info()
```

```python
def convert_sqft_to_num(x):
    x = str(x).strip()

    if "-" in x:
        tokens = x.split("-")
        try:
            return (float(tokens[0].strip()) + float(tokens[1].strip())) / 2
        except:
            return None

    try:
        return float(x)
    except:
        return None
```

```python
df["total_sqft"] = df["total_sqft"].apply(convert_sqft_to_num)
```

```python
df = df.dropna(subset=["total_sqft"])
```

```python
print(df["total_sqft"].dtype)
```

```python
df.to_csv("../data/raw/bengaluru_house_prices.csv", index=False)
```
